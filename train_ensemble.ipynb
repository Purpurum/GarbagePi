```python
from  ultralytics import YOLO
from collections import Counter
from PIL import Image
import yolov5
import os
import time
```

```python
# Load a model
model_variant = "yolov8n"
model = YOLO(f'{model_variant}.yaml') 

# Train the model
results = model.train(data='video2_data.yaml', epochs=100, imgsz=640, save_period=5, patience=10, workers=2, name="yolov8_video2_n")
```

```python
models = [YOLO("runs/detect/yolov8_video0_n/weights/best.pt"), 
          YOLO("runs/detect/yolov8_video1_n/weights/best.pt"), 
          YOLO("runs/detect/yolov8_video2_n/weights/best.pt"), 
          YOLO("runs/detect/yolov8_combined_pre_s/weights/best.pt"), 
          YOLO("runs/detect/yolov8_combined_s/weights/best.pt")]
results_list = []
```

```python
for model in models:
    results = model("datasets/yolo/yolofied-video0/images/test/video0-0043.png")
    names_count = {}
    names = model.names
    for name in names:
        names_count[name] = results[0].boxes.cls.tolist().count(name)
    results_list.append(names_count)
```

```python
results_list
```

```python
most_common_values = {}

for d in results_list:
    for key, value in d.items():
        if key in most_common_values:
            counts = most_common_values[key]
            if value in counts:
                counts[value] += 1
            else:
                counts[value] = 1
        else:
            most_common_values[key] = {value: 1}

for key, counts in most_common_values.items():
    most_common = max(counts, key=counts.get)
    print(most_common, end=" ")
```
